# Multivariate Analysis

```python
numeric_columns = df.select_dtypes(include=['int64','float64'])

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Plot the correlation matrix
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Numeric Housing Features')
plt.show()

# Select features that are most correlated with SalePrice
correlated_features = correlation_matrix['SalePrice'].sort_values(ascending=False).index[1:11]

# Plot pairplot for the top correlated features
sns.pairplot(numeric_columns[correlated_features], diag_kind='kde')
plt.suptitle('Scatterplot Matrix of Top Correlated Features', y=1.02)
plt.show()
```

# Feature Engineering

```python
# Engineer new features
# Calculate price per square foot
df['PricePerSqFt'] = df['SalePrice'] / df['GrLivArea']

# Calculate the property's age at the time of sale
df['PropertyAge'] = df['YrSold'] - df['YearBuilt']

# Check the new features
print(df[['SalePrice', 'GrLivArea', 'PricePerSqFt', 'YearBuilt', 'YrSold', 'PropertyAge']].head())
```

# Feature Engineering and Size Impact:

```python
key_features = ['BedroomAbvGr', 'FullBath', 'GrLivArea', 'SalePrice']

# Scatter plots to visualize relationships between key features and house prices
plt.figure(figsize=(15, 10))

for i, feature in enumerate(key_features[:-1]):
    plt.subplot(2, 2, i+1)
    sns.scatterplot(data=df, x=feature, y='SalePrice')
    plt.title(f'Relationship between {feature} and SalePrice')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')

plt.tight_layout()
plt.show()

# Calculate and display the correlation matrix for key features
correlation_matrix = df[key_features].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Key Features and SalePrice')
plt.show()
```

# Market Trends & Historical Pricing

```python
import matplotlib.pyplot as plt

# Aggregate data by month
df['month'] = df['date'].dt.to_period('M')
monthly_prices = df.groupby('month')['SalePrice'].mean().reset_index()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(monthly_prices['month'].astype(str), monthly_prices['SalePrice'])
plt.xlabel('Month')
plt.ylabel('Average Sale Price')
plt.title('Average House Prices Over Time')
plt.xticks(rotation=90)
plt.show()
```
